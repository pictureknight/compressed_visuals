```python
%%capture
import base64
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import distutils.util
import os

current = os.getcwd()
inside = os.listdir(current)
for item in inside:
    if item.endswith('.png'):
        os.remove(os.path.join(current, item))
inp = input()
whtspace = ' '.join(inp.split()).replace(' ', '')
splt1 = whtspace.split('src')                               # split on keyword 'src'
splt2 = str(splt1[1]).split('"]}],"')
temp = splt2[0]
remove = '\":[,'
for n in remove:
    temp = temp.replace(n, '')
byt = bytes(temp, 'utf-8')
decoded = base64.b64decode(byt.decode('utf-8'))
decodedStr = str(decoded, 'utf-8')
li = decodedStr.split('\\n')
head = li[0].split(',')
name = head[0].replace('\'', '').replace('[', '')
lnwidth = head[1].replace('\'', '').replace(' ', '')
lncolor = head[2].replace('\'', '').replace(' ', '')
bgcolor = head[3].replace('\'', '').replace(' ', '')
rainbow = head[4].replace('\'', '').replace(' ', '')
rainbow = bool(distutils.util.strtobool(rainbow))
period = int(head[5].replace('\'', ''))
N = int(head[6].replace('\'', '').replace(']', ''))
print(name, lnwidth, lncolor, bgcolor, rainbow, period, N)
for i in range(1, 4*N+1):
    li[i] = float(li[i])
pts=10000
timestamps = np.linspace(0, period, pts)
timestamps = np.array(timestamps)
real_coefs_x = np.array(li[1:N+1])
imag_coefs_x = np.array(li[N+1:2*N+1])
real_coefs_y = np.array(li[2*N+1:3*N+1])
imag_coefs_y = np.array(li[3*N+1:4*N+1])
coef_list_x = []
coef_list_y = []
for n in range(0, len(real_coefs_x)):
    coef_list_x.append(real_coefs_x[n]+1j*imag_coefs_x[n])
for n in range(0, len(real_coefs_y)):
    coef_list_y.append(real_coefs_y[n]+1j*imag_coefs_y[n])
def f(x, degree=N):
    f = np.array([2*coefs[i-1]*np.exp(1j*2*i*np.pi*x/period) for i in range(1,degree+1)])
    return(f.sum())
coefs = coef_list_x
fourier_approximation_x = np.array([f(t, degree=N).real for t in timestamps])
xmin = np.amin(fourier_approximation_x)
xmax = np.amax(fourier_approximation_x)
coefs = coef_list_y
fourier_approximation_y = np.array([f(t, degree=N).real for t in timestamps])
ymin = np.amin(fourier_approximation_y)
ymax = np.amax(fourier_approximation_y)
if not rainbow:
    plt.figure(figsize=(12,12))
    plt.gca().set_aspect('equal')
    plt.axis('off')
    buffer = 10
    plt.xlim(xmin - buffer, xmax + buffer)
    plt.ylim(ymin - buffer, ymax + buffer)
    plt.plot(fourier_approximation_x, fourier_approximation_y, linewidth=lnwidth, color=lncolor)
    plt.savefig(name+'.png', dpi=200, facecolor=bgcolor, edgecolor=bgcolor, transparent=True, bbox_inches=None, pad_inches=0.0)
elif rainbow:
    from matplotlib.collections import LineCollection
    plt.rcParams.update({'axes.facecolor':bgcolor})
    plt.rcParams['lines.linewidth'] = lnwidth
    plt.figure(figsize=(12,12))
    plt.gca().set_aspect('equal')
    plt.axis('off')
    x = fourier_approximation_x
    y = fourier_approximation_y
    t = np.linspace(0,1,x.shape[0])
    points = np.array([x,y]).transpose().reshape(-1,1,2)
    segs = np.concatenate([points[:-2], points[1:-1], points[2:]], axis=1)
    lc = LineCollection(segs, cmap=plt.get_cmap('jet'))
    lc.set_array(t)
    plt.gca().add_collection(lc)
    buffer = 10
    plt.xlim(x.min() - buffer, x.max() + buffer)
    plt.ylim(y.min() - buffer, y.max() + buffer)
    plt.savefig(name+'.png', dpi=200, facecolor=bgcolor, edgecolor=bgcolor, transparent=True, bbox_inches=None, pad_inches=0.0)


##########################################################################################################


###     PASTE THE METADATA IN THE FIELD BELOW, THEN PRESS ENTER     ###
```
